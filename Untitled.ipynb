```python
import numpy as np
import matplotlib.pyplot as plt

# Define grid size
size = 100
# Create a grid of distances from the central point
x = np.arange(size)
y = np.arange(size)
X, Y = np.meshgrid(x, y)

# Define the positions of PCBs in each corner of the rectangle
pcb_positions = [(0, 0), (0, size - 1), (size - 1, 0), (size - 1, size - 1)]

# Calculate the distance from each grid cell to each PCB
distances_to_pcbs = np.zeros((size, size, len(pcb_positions)))
for i, (pcb_x, pcb_y) in enumerate(pcb_positions):
    distances_to_pcbs[:, :, i] = np.sqrt((X - pcb_x)**2 + (Y - pcb_y)**2)

# Combine distances to get the total distance from each grid cell to the case
total_distance_to_case = np.sum(distances_to_pcbs**2, axis=2)**0.5

# Stefan-Boltzmann constant
sigma = 5.67e-8  # W/m^2/K^4

# Temperature calculation (for simplicity, assume surroundings temperature is 0K)
Ts = 300  # Surface temperature in Kelvin
T_surroundings = 0  # Surroundings temperature in Kelvin
epsilon = 1  # Assume perfect emitter

# Calculate the temperature based on the Stefan-Boltzmann Law
temperature = Ts**4 - (total_distance_to_case * sigma * epsilon * (Ts**4 - T_surroundings**4))

data = np.random.rand(10, 10)


plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar(label='Value')

# Define the coordinates of the box (in data coordinates)
box_x = [2, 7]  # x-coordinates of the box (start and end)
box_y = [3, 8]  # y-coordinates of the box (start and end)

# Plot the rectangular box
plt.plot([box_x[0], box_x[1], box_x[1], box_x[0], box_x[0]], 
         [box_y[0], box_y[0], box_y[1], box_y[1], box_y[0]], 
         color='blue', linewidth=2)


# Create the heatmap
plt.imshow(temperature, cmap='hot', interpolation='nearest')
plt.colorbar(label='Temperature (K)')
plt.title('Radiation heatmap from the CPU using boltzmans law ')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```
